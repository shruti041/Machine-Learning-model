```python
#Before going further, it is extremely important to know if our data contains any null values. Null values affects the performance and accuracy of the machine learning model and therefore it is important to identify them and deal with them accordingly.
data.isnull().sum()
```

```python
#we will replace the null values with mean values for each respective column because if we delete the column we lost alot of data.
columns_with_null_values = ['cigsPerDay','BPMeds','totChol','BMI','heartRate','glucose']
for i in columns_with_null_values: 
    data[i].fillna(data[i].mean(),inplace=True)
#Check again null values
data.isnull().sum()
```

```python
sns.pairplot(data)
plt.show()
```

```python
#The statistical relationship between two variables is referred to as their correlation. 
# When two features have high correlation it simply means that they have a strong relationship with each other.
# For better performance and accuracy of the machine learning models, it is highly recommended to drop all features that have high correlation.
# highly correlated features increase the complexity of the algorithm, thus increasing the risk of errors.
# A correlation heatmap is a graphical tool that displays the correlation between multiple variables as a color-coded matrix. 
# From the seaborn heatmap below, we can identify the highly correlated features by looking at the intensity of colours for each grid. 
# The more lighter the colour the more the correlation between respective features. 
# The diagonal of the matrix looks highly correlated since along the diagonal, each feature is plotted against itself.
```

```python
sns.heatmap(data.corr())
plt.show()
```

```python
# Identify and remove highly correlated features
highly_correlated_features = ['currentSmoker','diaBP','prevalentHyp','diabetes']
# remove these features from the dataset since they highly correlate with other features from the dataset
data.drop(highly_correlated_features,axis=1,inplace=True)
#Plot the correlation heatmap again to confirm
sns.heatmap(data.corr())
plt.show()
```

```python
df = pd.DataFrame(data)      #DataFrame is a 2-dimensional labeled data structure 
print(df) 
```

```python
#countplot of dataset
TyCHD=sns.countplot(x='TenYearCHD',data=df,hue='TenYearCHD')
for bars in TyCHD.containers:
    TyCHD.bar_label(bars)
plt.show()
#It contains about 3594 patients without the risk and only around 644 patients identified under risk.
```

## Prepare dataset for training

```python
#Now that we have normalized the null values and removed the highly correlated features. We are ready to build and train our model.
```

```python
# we will treat the 'TenYearCHD' as our output feature (y).
# The rest of the features will be treated as the predictors (X).
X = data.drop('TenYearCHD',axis=1)
y = data['TenYearCHD']
```

```python
#Split dataset to train and test sets
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

```python
X_train.head()
```

```python
#Scale the data before setting them to the model
#Different scales of the data features affect the modeling of a dataset adversely.
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=pd.DataFrame(sc.fit_transform(X_train))
X_test=pd.DataFrame(sc.transform(X_test))
X_train.head()
```

```python
#Scaling removed column indexes
X_train.columns= X.columns   
X_test.columns= X.columns

y_train.index= X_train.index
y_test.index= X_test.index

X_train.head()
```

## Logistic regression model 

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Make predictions on the test set
predictions=model.predict(X_test)
```

```python
#Custom function to print the confusion matrix 
from sklearn.metrics import classification_report
def my_confusion_matrix(y_test, y_pred, plt_title):
    cm=confusion_matrix(y_test, y_pred)
    print(classification_report(y_test, y_pred))
    sns.heatmap(cm, annot=True, fmt='g', cbar=False, cmap='BuPu')
    plt.xlabel('Predicted Values')
    plt.ylabel('Actual Values')
    plt.title(plt_title)
    plt.show()
    return cm
```

```python
#Evaluate the model
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Accuracy with Logistic Regression: ',accuracy_score(y_test,predictions))
```

```python
print('The confusion matrix',my_confusion_matrix(y_test,predictions,'Logistic Regression CM'))
```

## K-Nearest Neighbors (KNN) model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Visualisation for the Error Rate/K-value 
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i, metric = 'minkowski', p=2)
    knn.fit(X_train,y_train)
    pred_i_knn = knn.predict(X_test)
    error_rate.append(np.mean(pred_i_knn != y_test))
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
#Notice that the lowest error rate is at k=17
#Fitting the KNN model
knn_classifier = KNeighborsClassifier(n_neighbors = 17, metric = 'minkowski', p=2)   #standard Euclidean distance when p = 2. 
knn_classifier.fit(X_train, y_train)
y_pred_knn=knn_classifier.predict(X_test)
print('KNN Accuracy Score: ',accuracy_score(y_test,y_pred_knn))
```

```python
cm_knn=my_confusion_matrix(y_test, y_pred_knn, 'KNN Confusion Matrix')
```
